```python
@run(scope='ospray', after=print)
def world(lib, /):
    @auto.contextlib.contextmanager
    def commit(obj):
        try:
            yield obj
        
        finally:
            lib.ospCommit(obj)
            
    #/commit
    C = commit
    
    @auto.contextlib.contextmanager
    def release(obj):
        try:
            yield obj
        
        finally:
            lib.ospRelease(obj)
            
    #/release
    R = release
    
    def data(type: lib.OSPDataType, arr: auto.numpy.ndarray, /) -> lib.OSPData:
        if len(arr.shape) == 0:
            arr = arr[None, None, None]
        
        elif len(arr.shape) == 1:
            arr = arr[:, None, None]
        
        elif len(arr.shape) == 2:
            arr = arr[:, :, None]
        
        elif len(arr.shape) == 3:
            arr = arr[:, :, :]
        
        else:
            raise NotImplementedError()

        with R( lib.ospNewSharedData(
            auto.ctypes.c_void_p( arr.ctypes.data ),
            type,
            arr.shape[0], 0,
            arr.shape[1], 0,
            arr.shape[2], 0,
        ) ) as src:
            with C( lib.ospNewData(
                type,
                arr.shape[0],
                arr.shape[1],
                arr.shape[2],
            ) ) as dst:
                lib.ospCopyData(
                    src,
                    dst,
                    0,
                    0,
                    0,
                )

            #/dst
        #/src
    
        return dst
    
    #/data
    D = data
    
    def pointer(x):
        return auto.ctypes.cast(x, auto.ctypes.c_void_p).value
    
    #/pointer
    P = pointer
    
    count = auto.itertools.count()
    with auto.contextlib.ExitStack() as stack:
        E = stack.enter_context

        with C( lib.ospNewWorld() ) as world:
            with C( lib.ospNewInstance(None) ) as instance:
                with C( lib.ospNewGroup() ) as group:
                    with C( lib.ospNewGeometricModel(None) ) as model:
                        with C( lib.ospNewGeometry(b'mesh') ) as geometry:
                            lib.ospSetParam(
                                geometry, b'vertex.position', lib.OSP_DATA,
                                auto.ctypes.byref( E(R(D(
                                    lib.OSP_VEC3F,
                                    auto.numpy.fromfile(
                                        'data/OSPGeometry.mesh.vertex.position.vec3f.bin',
                                        dtype='f4,f4,f4',
                                    ),
                                ))) ),
                            )

                            lib.ospSetParam(
                                geometry, b'index', lib.OSP_DATA,
                                auto.ctypes.byref( E(R(D(
                                    lib.OSP_VEC4UI,
                                    auto.numpy.fromfile(
                                        'data/OSPGeometry.mesh.index.vec4ui.bin',
                                        dtype='u4,u4,u4,u4',
                                    ),
                                ))) ),
                            )
                            
                        #/geometry
                        lib.ospSetParam(
                            model, b'geometry', lib.OSP_GEOMETRY,
                            auto.ctypes.byref( E(R(geometry)) ),
                        )
                        
                        with C( lib.ospNewMaterial(None, b'obj') ) as material:
                            lib.ospSetParam(
                                material, b'kd', lib.OSP_VEC3F,
                                (auto.ctypes.c_float * 3)(0.7, 0.3, 0.7),
                            )
                            
                        #/material
                        lib.ospSetParam(
                            model, b'material', lib.OSP_MATERIAL,
                            auto.ctypes.byref( E(R(material)) ),
                        )
                        
                    #/model
                    lib.ospSetParam(
                        group, b'geometry', lib.OSP_DATA,
                        auto.ctypes.byref( E(R(D(
                            lib.OSP_GEOMETRIC_MODEL,
                            auto.numpy.array([
                                P(model),
                            ], dtype=auto.numpy.uintp),
                        ))) ),
                    )
                    
                #/group
                lib.ospSetParam(
                    instance, b'group', lib.OSP_GROUP,
                    auto.ctypes.byref( E(R(group)) ),
                )

            #/instance
            lib.ospSetParam(
                world, b'instance', lib.OSP_DATA,
                auto.ctypes.byref( E(R(D(
                    lib.OSP_INSTANCE,
                    auto.numpy.array([
                        P(instance),
                    ], dtype=auto.numpy.uintp),
                ))) ),
            )
            
        #/world
    #/stack
    return world
```

```python
@run(scope='ospray', after=print)
def __render(lib, world, /):
    @auto.contextlib.contextmanager
    def commit(obj):
        try:
            yield obj
        
        finally:
            lib.ospCommit(obj)
            
    #/commit
    C = commit
    
    @auto.contextlib.contextmanager
    def release(obj):
        try:
            yield obj
        
        finally:
            lib.ospRelease(obj)
            
    #/release
    R = release
    
    def data(type: lib.OSPDataType, arr: auto.numpy.ndarray, /) -> lib.OSPData:
        if len(arr.shape) == 0:
            arr = arr[None, None, None]
        
        elif len(arr.shape) == 1:
            arr = arr[:, None, None]
        
        elif len(arr.shape) == 2:
            arr = arr[:, :, None]
        
        elif len(arr.shape) == 3:
            arr = arr[:, :, :]
        
        else:
            raise NotImplementedError()

        with R( lib.ospNewSharedData(
            auto.ctypes.c_void_p( arr.ctypes.data ),
            type,
            arr.shape[0], 0,
            arr.shape[1], 0,
            arr.shape[2], 0,
        ) ) as src:
            with C( lib.ospNewData(
                type,
                arr.shape[0],
                arr.shape[1],
                arr.shape[2],
            ) ) as dst:
                lib.ospCopyData(
                    src,
                    dst,
                    0,
                    0,
                    0,
                )

            #/dst
        #/src
    
        return dst
    
    #/data
    D = data
    
    def pointer(x):
        return auto.ctypes.cast(x, auto.ctypes.c_void_p).value
    
    #/pointer
    P = pointer
    
    @auto.mediocreatbest.immediate
    @auto.mediocreatbest.coroutine
    def display():
        handle = auto.IPython.display.display(None, display_id=True)
        
        while True:
            obj = yield
            handle.update(obj)
    
    count = auto.itertools.count()
    with auto.contextlib.ExitStack() as stack:
        E = stack.enter_context
        
        with C( lib.ospNewRenderer(b'ao') ) as renderer:
            pass
        
        #/renderer
        
        with C( lib.ospNewCamera(b'perspective') ) as camera:
            lib.ospSetParam(
                camera, b'position', lib.OSP_VEC3F,
                (auto.ctypes.c_float * 3)(
                    577.104925932147, 3708.200840427869, -5150.856766744237
                ),
            )
            
            lib.ospSetParam(
                camera, b'up', lib.OSP_VEC3F,
                (auto.ctypes.c_float * 3)(
                    577.104925932147, 3708.200840427869, -5150.856766744237
                ),
            )
            
            lib.ospSetParam(
                camera, b'direction', lib.OSP_VEC3F,
                (auto.ctypes.c_float * 3)(
                    3.210875083447263, 0.0941845222973825, 1.0472723045186285
                ),
            )
        
        #/camera
        
        with R( lib.ospNewFrameBuffer((w := 512), (h := 512), lib.OSP_FB_RGBA8, lib.OSP_FB_COLOR) ) as frame:
            _: float = lib.ospRenderFrameBlocking(
                frame,
                renderer,
                camera,
                world,
            )
            
            rgba = lib.ospMapFrameBuffer(frame, lib.OSP_FB_COLOR)
            image = auto.PIL.Image.frombytes(
                'RGB',
                (w, h),
                auto.ctypes.string_at(rgba, size=(w * h * 4)),
                'raw',
                'RGBX',
                0,
                -1,
            )
#             image = image.convert('RGB')
            
            display.send(image)
            
            lib.ospUnmapFrameBuffer(rgba, frame)

        #/frame
        
    #/stack
```

```python
def convert_location_to_spatial(
    *,
    lat: Degree,
    lng: Degree,
    alt: Meter,
    math=auto.math,
) -> Tuple[Meter, Meter, Meter]:
    # Thanks https://gis.stackexchange.com/a/4148
    
    #> Note that "Lat/Lon/Alt" is just another name for spherical coordinates, and 
    #> phi/theta/rho are just another name for latitude, longitude, and altitude.
    #> :) (A minor difference: altitude is usually measured from the surface of the 
    #> sphere; rho is measured from the center -- to convert, just add/subtract the 
    #> radius of the sphere.)
    phi: Radian = math.radians(lat)
    theta: Radian = math.radians(lng)
    
    # Thanks https://en.wikipedia.org/wiki/Earth_radius
    #> A globally-average value is usually considered to be 6,371 kilometres (3,959 mi)
    rho: Meter = 6_371_000 + alt
    
    #> x = math.cos(phi) * math.cos(theta) * rho
    x: Meter = math.cos(phi) * math.cos(theta) * rho
    
    #> y = math.cos(phi) * math.sin(theta) * rho
    y: Meter = math.cos(phi) * math.sin(theta) * rho

    #> z = math.sin(phi) * rho # z is 'up'
    z: Meter = math.sin(phi) * rho
    
    #> (Note there's some slightly arbitrary choices here in what each axis means...
    #> you might want 'y' to point at the north pole instead of 'z', for example.)
    
    # I do :)
    y, z = z, y
    
    return x, y, z

@auto.dataclasses.dataclass(eq=True, frozen=True)
class View:
    position: tuple[float, float, float]
    up: tuple[float, float, float]
    direction: tuple[float, float, float]
        
    def __post_init__(self):
        def unit(v: auto.numpy.typing.ArrayLike):
            v = auto.numpy.asarray(v)
            v = v / auto.numpy.linalg.norm(v)
            return v
        
        unit_up = unit(self.up)
        unit_direction = unit(self.direction)
        assert (dot := auto.numpy.dot(unit_up, unit_direction) < 0.99), \
            f"""Camera up and direction are too similar: {dot=!r}"""

@run(after=lambda x: print(x[len(x)//2]))
def views() -> list[View]:
    with open(auto.pathlib.Path.cwd() / 'data' / 'Appalachian_Trail.json', 'r') as f:
        trail = auto.json.load(f)

#         trail = trail[::8]

    views = []
    for window in auto.more_itertools.sliding_window(trail, 3 * (k := 4)):
        srcs = window[:k]
        tgts = window[-k:]
        assert len(srcs) == k
        assert len(tgts) == k

        src_lat = auto.statistics.fmean(src["lat"] for src in srcs)
        src_lng = auto.statistics.fmean(src["lng"] for src in srcs)
        src_alt: Meter = 2_000 # * auto.random.uniform(0.9, 1.1)

        tgt_lat = auto.statistics.fmean(tgt["lat"] for tgt in tgts)
        tgt_lng = auto.statistics.fmean(tgt["lng"] for tgt in tgts)
        tgt_alt: Meter = 1_500 # * auto.random.uniform(0.9, 1.1)

        src_x, src_y, src_z = map(lambda x: x/1000, convert_location_to_spatial(
            lat=src_lat,
            lng=src_lng,
            alt=src_alt,
        ))

        tgt_x, tgt_y, tgt_z = map(lambda x: x/1000, convert_location_to_spatial(
            lat=tgt_lat,
            lng=tgt_lng,
            alt=tgt_alt,
        ))

        views.append(View(
            position=(
                src_x,
                src_y,
                src_z,
            ),
            up=(
                src_x,
                src_y,
                src_z,
            ),
            direction=(
                tgt_x - src_x,
                tgt_y - src_y,
                tgt_z - src_z,
            ),
        ))

    return views
```
